```python
# Imports packages
import random, os, sys, shutil
import cv2
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np
import keras
```

```python
# Check if there are unreadable images in the source folder
filenames= os.listdir("C:/Users/yaoji/Desktop/Jing/Good")
for file in filenames:
    try:
        image = cv2.resize(cv2.imread(file), (224, 224))
    except:
        print(file)
```

```python
class image_clustering:

    # define key parameters
    def __init__(self, folder_path="data", n_clusters=3, max_examples=None):
        paths = os.listdir(folder_path)
        if max_examples == None:          # max_examples gives the options to take how many images into clustering criteria
            self.max_examples = len(paths)
        else:
            if max_examples > len(paths):
                self.max_examples = len(paths)
            else:
                self.max_examples = max_examples
        self.n_clusters = n_clusters
        self.folder_path = folder_path
        random.shuffle(paths)             # Modify a sequence in-place by shuffling its contents. 
        self.image_paths = paths[:self.max_examples]
        del paths 
        try:
            shutil.rmtree("output")       # You may need to create a folder named "output" yourself
        except FileExistsError:
            pass
        print("\n output folders created.")
        os.makedirs("output")
        for i in range(self.n_clusters):  # In this step, subfolders named cluster0, cluster1, cluster2,...will be created under "output" folder
            os.makedirs("output\\cluster" + str(i))
        print("\n Object of class \"image_clustering\" has been initialized.")

    #Load images
    def load_images(self):
        self.images = []
        for image in self.image_paths:
            self.images.append(cv2.cvtColor(cv2.resize(cv2.imread(self.folder_path + "\\" + image), (224,224)), cv2.COLOR_BGR2RGB))
            # Here resize is in case we want to furthur improve the code by adding some NN in it. No need to resize if not using NN.
        self.images = np.float32(self.images).reshape(len(self.images), -1)
        self.images /= 255
        print("\n " + str(self.max_examples) + " images from the \"" + self.folder_path + "\" folder have been loaded in a random order.")

    #Perform Kmeans clustering
    def clustering(self):
        model = KMeans(n_clusters=self.n_clusters, n_jobs=-1, random_state=728)
        self.images_new = self.images
        model.fit(self.images_new)
        predictions = model.predict(self.images_new)
        #print(predictions)
        for i in range(self.max_examples):
            shutil.copy2(self.folder_path+"\\"+self.image_paths[i], "output\cluster"+str(predictions[i]))
        print("\n Finish Clustering \n\n Clusters and the respective images are stored in the \"output\" folder.")

# when this module is operating directly as main program, the below code will be executed
# If this modulse is imported, it will not be executed
if __name__ == "__main__":
    print("Executing as main program")
    print("Value of __name__ is: ", __name__)
    print("\n\n \t\t Clustering START\n\n")
    
    number_of_clusters = 3                          # from cluster 0 to cluster n-1
    data_path = "C:/Users/yaoji/Desktop/Jing/Good"  # path of the image folder (Must contain only images)
    max_examples = None                             # number of examples to use, "None" means all images will be taken into consideration for clustering 

# call previously defined function
    temp = image_clustering(data_path, number_of_clusters, max_examples) 
    temp.load_images()
    temp.clustering()

    print("\n\n\t\t Clustering END\n\n")
```
